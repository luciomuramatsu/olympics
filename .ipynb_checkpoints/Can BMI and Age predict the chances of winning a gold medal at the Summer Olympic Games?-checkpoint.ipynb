```python
# drop all the entries that do not have values for Height, Weight and Age
olympic_cleaned_1960_2016 = pysqldf('''
SELECT 
    *
FROM olympics_summer_1960_2016 
WHERE 
    Height IS NOT NULL
    AND
    Weight IS NOT NULL
    AND 
    Age IS NOT NULL
;
''')
```

## 5 - Data Binning

Since every sport has its own strategy, tactic, and physical demand. In the next steps, I will split the dataset into 3 groups based on:

- Team or individual sports
- Body-Weight classified sports

```python
# Get a list of all the sports
all_sports = olympic_cleaned_1960_2016.Sport.unique().tolist()
all_sports
```

#### Groups 

1. Team sports
2. Weight-class sports
3. Individual sports

*note that some sports may have both categories; individual and couple/teams. I've opted to keep in the team sports group only the sports that are played exclusively in teams.

```python
team_sports = ['Football', 'Hockey', 'Basketball', 'Handball', 'Volleyball', 'Water Polo',
               'Synchronized Swimming', 'Baseball', 'Badminton', 'Softball', 'Rugby Sevens',
              'Beach Volleyball']
weight_sports = ['Wrestling', 'Boxing', 'Judo', 'Taekwondo', 'Weightlifting']
individual_sports = ['Athletics', 'Swimming', 'Gymnastics', 'Rowing', 'Cycling', 'Shooting', 'Fencing', 
                     'Canoeing', 'Sailing', 'Equestrianism', 'Diving', 'Archery', 'Tennis', 'Triathlon', 
                     'Modern Pentathlon', 'Table Tennis', 'Rhythmic Gymnastics', 'Trampolining', 'Golf']
```

#### Creating new variables.

1. Body Mass Index (BMI) = Weight / Height  2 -> type float
2. medalist (Won any medal = True, No medal = False)  -> type boolean
3. medal_points (Gold = 3, Silver = 2, Bronze = 1, no medal = 0) -> type int

```python
olympic_cleaned = pysqldf('''
SELECT 
    Id,
    Sex,
    Team,
    Age,
    Height,
    Weight,
    (weight / ((height/100)*(height/100))) AS BMI,
    Year,
    Sport,
    Event,
    CASE 
        WHEN Medal = 'Bronze' THEN 1
        WHEN Medal = 'Silver' THEN 2
        WHEN Medal = 'Gold' THEN 3
        ELSE 0
        END AS medal_points,
    CASE 
        WHEN Medal = 'Bronze' THEN 1
        WHEN Medal = 'Silver' THEN 1
        WHEN Medal = 'Gold' THEN 1
        ELSE 0
        END AS medalist
        
FROM olympic_cleaned_1960_2016 
;
''')

olympic_cleaned.head()
```

```python
# Adding a new column to label the entry in one of the 3 groups (team, weight, and indiv).

olympic_cleaned.loc[:,'sport_group'] = olympic_cleaned.Sport.apply(lambda x: 'team' if x in team_sports else 
                                                             ('weight' if x in weight_sports else
                                                             'indiv'))
```

```python
# Checking the size of each group
olympic_cleaned.sport_group.value_counts()
```

## 6 - Analysis

- ANOVA
- Tukey Test

#### Histograms to check for normal distribution

```python
male_cleaned = olympic_cleaned[olympic_cleaned['Sex'] == 'M']
female_cleaned = olympic_cleaned[olympic_cleaned['Sex'] == 'F']
```

```python
sns.FacetGrid(male_cleaned, col="sport_group", row='medalist').map(plt.hist, 'BMI').add_legend()
```

```python
sns.FacetGrid(female_cleaned, col="sport_group", row='medalist').map(plt.hist, 'BMI').add_legend()
```

```python
# Shapiro-Wilk normality test 
print('M - Team - Medal')
print(stats.shapiro(male_cleaned[(male_cleaned['sport_group'] == 'team') & (male_cleaned['medalist'] == 1)]['BMI']))
print('M - Team - No')
print(stats.shapiro(male_cleaned[(male_cleaned['sport_group'] == 'team') & (male_cleaned['medalist'] == 0)]['BMI']))
print('F - Team - Medal')
print(stats.shapiro(female_cleaned[(female_cleaned['sport_group'] == 'team') & (female_cleaned['medalist'] == 1)]['BMI']))
print('F - Team - No')
print(stats.shapiro(female_cleaned[(female_cleaned['sport_group'] == 'team') & (female_cleaned['medalist'] == 0)]['BMI']))
# alpha level = 0.05. If pvalue > 0.05, the group is normally distributed.
```

```python
# Shapiro-Wilk normality test 
print('M - Weight - Medal')
print(stats.shapiro(male_cleaned[(male_cleaned['sport_group'] == 'weight') & (male_cleaned['medalist'] == 1)]['BMI']))
print('M - Weight - No')
print(stats.shapiro(male_cleaned[(male_cleaned['sport_group'] == 'weight') & (male_cleaned['medalist'] == 0)]['BMI']))
print('F - Weight - Medal')
print(stats.shapiro(female_cleaned[(female_cleaned['sport_group'] == 'weight') & (female_cleaned['medalist'] == 1)]['BMI']))
print('F - Weight - No')
print(stats.shapiro(female_cleaned[(female_cleaned['sport_group'] == 'weight') & (female_cleaned['medalist'] == 0)]['BMI']))
# alpha level = 0.05. If pvalue > 0.05, the group is normally distributed.
```

```python
# Shapiro-Wilk normality test 
print('M - indiv - Medal')
print(stats.shapiro(male_cleaned[(male_cleaned['sport_group'] == 'indiv') & (male_cleaned['medalist'] == 1)]['BMI']))
print('M - indiv - No')
print(stats.shapiro(male_cleaned[(male_cleaned['sport_group'] == 'indiv') & (male_cleaned['medalist'] == 0)]['BMI']))
print('F - indiv - Medal')
print(stats.shapiro(female_cleaned[(female_cleaned['sport_group'] == 'indiv') & (female_cleaned['medalist'] == 1)]['BMI']))
print('F - indiv - No')
print(stats.shapiro(female_cleaned[(female_cleaned['sport_group'] == 'indiv') & (female_cleaned['medalist'] == 0)]['BMI']))
# alpha level = 0.05. If pvalue > 0.05, the group is normally distributed.
```

```python
# Create new columns to label the experiment groups
olympic_cleaned.loc[:,'medalistx'] = olympic_cleaned.loc[:,'medalist'].replace({0: '0', 1: '1'})
olympic_cleaned.loc[:,'Group'] = olympic_cleaned.loc[:,'Sex'] + '-' + olympic_cleaned.loc[:,'sport_group'] + '-' + olympic_cleaned.loc[:,'medalistx']
```

```python
sns.FacetGrid(olympic_cleaned, row="medalist", col='Sex').map(sns.boxplot, x = olympic_cleaned.BMI, y = olympic_cleaned.sport_group).add_legend()
```

```python
male_cleaned = olympic_cleaned[olympic_cleaned['Sex'] == 'M']
female_cleaned = olympic_cleaned[olympic_cleaned['Sex'] == 'F']
```

```python
# One-way ANOVA to test equality between groups means.
# if pvalue < 0.05. The group means are different
medal_team = (male_cleaned['medalist'] == 1) & (male_cleaned['sport_group'] == 'team')
no_team = (male_cleaned['medalist'] == 0) & (male_cleaned['sport_group'] == 'team')

medal_weight = (male_cleaned['medalist'] == 1) & (male_cleaned['sport_group'] == 'weight')
no_weight = (male_cleaned['medalist'] == 0) & (male_cleaned['sport_group'] == 'weight')

medal_indiv = (male_cleaned['medalist'] == 1) & (male_cleaned['sport_group'] == 'indiv')
no_indiv = (male_cleaned['medalist'] == 0) & (male_cleaned['sport_group'] == 'indiv')

F, p = stats.f_oneway(male_cleaned[medal_team]['BMI'], 
                      male_cleaned[no_team]['BMI'],
                      male_cleaned[medal_weight]['BMI'],
                      male_cleaned[no_weight]['BMI'],
                      male_cleaned[medal_indiv]['BMI'],
                      male_cleaned[no_indiv]['BMI'])
print('F statistic = {:.10f}'.format(F))
print('p value = {:.10f}'.format(p))
```

```python
# Turkey test to compare groups
mcobj = ml.MultiComparison(male_cleaned['BMI'], male_cleaned['Group'])
out = mcobj.tukeyhsd(0.05) 
print('Table 1 - Multicomparison between All Male groups ')
print(out)
```

```python
# Turkey test to compare groups
mcobj = ml.MultiComparison(female_cleaned['BMI'], female_cleaned['Group'])
out = mcobj.tukeyhsd(0.05) 
print('Table 2 - Multicomparison between All Male groups ')
print(out)
```

## 9 - Discussion and Conclusion

Age and BMI were a statistically significant factor for performance among Olympic athletes.
In the majority of the sports, optimal age and BMI were concentrated in:

- Age: between 20 and 35 years
- BMI: between 18 and 25 kg/m2

Some sports, e.g., Gymnastics and Rhythmic Gymnastics, tend to have younger winners. On the other hand, sports like Equestrianism, Saling, and Shooting have a more disperse winning group age.

BMI also weight differently in some sports. Weight-classified sports, e. g., Judo, Boxing, Wrestling, and Weightlifting, have winners with BMI over 30 kg/m2.

In conclusion, body measurements such as age, weight, and height can be used to determine the chances of one being placed on the winning podium in the Summer Olympic Games.
